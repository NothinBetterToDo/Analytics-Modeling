Use case scenario: We will need to increase ROI for our sales department and based on the historical data: 

- Sales employee complete on average 50 calls per day
- Prob of conversion for each call is 4%
- Average revenue for each conversion is \$20
- Sales Department has 100 employees 
- Each employee is paid <span>$200</span> per day 

Assume each employee as a binomially distributed random variable with:
- n = 50
- p = 4%

```python
# libraries
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Simulation

# number of employees
e = 100

# cost per employee
w = 200 

# number of independent calls per employee
n = 50

# prob of success for each call 
p = 0.04

# revenue per call
r = 100

# binomial random variables of sales employess
conversions = np.random.binomial(n, p, size = e)

# key metrics 
print('Average Conversions per Employee:', 
      str(round(np.mean(conversions),2)))
print('Standard Deviation of Conversions per Employee:',
      str(round(np.std(conversions), 2)))
print('Total Conversions:', str(np.sum(conversions)))
print('Total Revenues:', str(np.sum(conversions)*r))
print('Total Expenses:', str(e*w))
print('Total Profits:', str((np.sum(conversions)*r) - (e * w)))
```

```python
# number of days to stimulate
sims = 1000

# binomial random variables of sales employess
sim_conversions = [np.sum(np.random.binomial(n, p, size = e)) for 
                   i in range(sims)]
sim_profits = np.array(sim_conversions)*r - e*w

# plot and save the results as a hist
fig, ax = plt.subplots(figsize = (14,7))
ax = sns.distplot(sim_profits, bins=20, label='Simulation Results')

ax.set_xlabel('Profits', fontsize=16)
ax.set_ylabel('Frequency', fontsize=16)

plt.savefig(fname='Hist', dpi=150)
```

```python
# negative profit
sum([1 for i in sim_profits if i < 0])
```

```python
# Simulation (with higher conversion rate 5%)

# number of employees
e = 100

# cost per employee
w = 200 

# number of independent calls per employee
n = 50

# prob of success for each call 
p = 0.05

# revenue per call
r = 100

# binomial random variables of sales employess
conversions = np.random.binomial(n, p, size = e)

# key metrics 
print('Average Conversions per Employee:', 
      str(round(np.mean(conversions),2)))
print('Standard Deviation of Conversions per Employee:',
      str(round(np.std(conversions), 2)))
print('Total Conversions:', str(np.sum(conversions)))
print('Total Revenues:', str(np.sum(conversions)*r))
print('Total Expenses:', str(e*w))
print('Total Profits:', str((np.sum(conversions)*r) - (e * w)))
```

```python
# number of days to stimulate
sims = 1000

# binomial random variables of sales employess
sim_conversions_up = [np.sum(np.random.binomial(n, p, size = e)) for 
                   i in range(sims)]
sim_profits_up = np.array(sim_conversions_up)*r - (e*w)

# plot and save the results as a hist
fig, ax = plt.subplots(figsize = (14,7))
ax = sns.distplot(sim_profits, bins=20, label='Original Simulation')
ax = sns.distplot(sim_profits_up, bins=20, label='Improved Simulation',
                 color='red')

ax.set_xlabel('Profits', fontsize=16)
ax.set_ylabel('Frequency', fontsize=16)
plt.legend()

plt.savefig(fname='Hist_Comparison', dpi=150)
```

```python
# negative profit for improved simulation 
sum([1 for i in sim_profits_up if i < 0])
```
