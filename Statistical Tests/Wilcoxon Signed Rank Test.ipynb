Wilcoxon Signed Rank Test is one of non-parametric statistical hypothesis test and can be used to compare paired data from the same population. 

```python
# packages 
from numpy.random import seed 
from numpy.random import randn
from scipy.stats import wilcoxon
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2 datasets
data1 = [1,5,4,5,5,9,5,5,0,0,5,0,6,6,0,0,0,1,2,2]
data2 = [7,7,10,9,8,3,2,2,10,10,6,10,10,6,10,5,10,8,9,10]

# print(len(data1))
# print(len(data2))
```

```python
# plot the distribution of the 2 data
sns.distplot(data1)
sns.distplot(data2)
```

```python
# compare datasets
stat, p = wilcoxon(data1, data2)
print('W = %.3f, p = %.3f' % (stat, p))
```

$$H_{0}: $$ <center>The median of the 2 datasets are the same.</center> <br>

$$H_{1}: $$ <center>The median of the 2 datasets are different.</center>

```python
alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis. Two medians are not the same.")
else:
    print("Fail to reject the null hypothesis. Two medians are the same.")
```
